## Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

### This task will be done in following steps:

- 1. First get the webpage https://www.naukri.com/
- 2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
- 3. Then click the searchbutton.
- 4. Then scrape the data for the first 10 jobs results you get.
- 5. Finally create a dataframe of the scraped data.

**Note: All of the above steps have to be done in code. No step is to be done manually.**

```python
# Connecting to driver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# opening the website
driver.get("https://www.naukri.com/")
```

```python
skill=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input")
skill.send_keys('Data Analyst')
```

```python
loc=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
loc.send_keys('Bangalore')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_designation=[]
job_location=[]
company_name=[]
exp_required=[]
```

```python
# scraping Job titile
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_designation.append(title)
    
# Scraping Job location
loc_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in loc_tags[0:10]:
    loc=i.text
    job_location.append(loc)

# Scraping company name
company=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company[0:10]:
    com=i.text
    company_name.append(com)

# Scraping Experience Required.
Experience=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in Experience[0:10]:
    exp=i.text
    exp_required.append(exp)
    
```

```python
print(len(job_designation),len(job_location),len(company_name),len(exp_required))
```

```python
df=pd.DataFrame({'Job Title':job_designation,'Job Location':job_location,'Company Name':company_name,'Experience Required':exp_required})
df
```

```python
driver.close()
```

## Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. ProductDescription
3. Price

The attributes which you have to scrape is ticked marked in the below image.



To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url :https://www.flipkart.com/
2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and 
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the required data as usual
4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100 sunglasses.

**Note: That all of the above steps have to be done by coding only and not manually.**


```python
# Connecting to driver
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# opening the website
driver.get("https://www.flipkart.com/")
```

```python
# entering sunglasses
skill=driver.find_element(By.CLASS_NAME,"_3704LK")
skill.send_keys('sunglasses')
```

```python
#  Clicking for search
search=driver.find_element(By.CLASS_NAME,'_34RNph')
search.click()
```

```python
brand=[]
product=[]
price=[]
discount=[]
```

```python
# loop for fetching the data from 3 page
j=0
for j in range(0,3):
    # scraping brand name
    brnd_name=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brnd_name:
        brnd=i.text
        brand.append(brnd)

    # Scraping product description
    prod=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in prod:
        prdt=i.text
        product.append(prdt)

    # Scraping product price
    prce=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in prce:
        pri=i.text
        price.append(pri)

    # Scraping discount percentage.
    discunt=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in discunt:
        dis=i.text
        discount.append(dis)
# Clicking for the next page         
search=driver.find_element(By.CLASS_NAME,'_1LKTO3')
search.click()
j+=1
```

```python
print(len(brand),len(product),len(price),len(discount))
```

```python
# DataFrame for 100 sunglasses
df=pd.DataFrame({'Brand':brand[0:100],'Product Discription':product[0:100],'Price':price[0:100],'Discount':discount[0:100]})
df
```

1. First get the webpagehttps://www.azquotes.com/

3. Than scrap 

a) Quote

b) Author

c) Type Of Quotes

```python
# Connecting the driver to crome
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# opening the website
driver.get("https://www.azquotes.com/")
```

```python
# Searching for TopQuotes
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
search.click()
```

```python
Quote=[]
Author=[]
Type_of_Quotes=[]
```

```python
# Fetching data from all the pages
j=0
for j in range(0,10):
    #scraping Quotes
    quo=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quo:
        quot=i.text
        Quote.append(quot)
    
    # Scraping Author name
    auth=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in auth:
        au=i.text
        Author.append(au)
    
    # Scraping Type of Quotes
    types=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in types:
        typ=i.text
        Type_of_Quotes.append(typ)
        
next=driver.find_elements(By.XPATH,'//li[@class="next"]')
j+=10
```

```python
print(len(Quote),len(Author),len(Type_of_Quotes))
```
